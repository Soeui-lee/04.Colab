```python
import numpy as np 
import matplotlib.pyplot as plt

plt.rc('figure', figsize=(12,8))
```

```python
fig = plt.figure()
```

```python
ax1 = fig.add_subplot(2, 2, 1) 
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3) 
ax3
```

```python
# 만든 도화지를 지정해서 그릴 수 있다. 

from numpy.random import randn
ax3.plot(randn(50).cumsum(), 'k--') # 무작위 50개의 수의 누적합 
_ = ax1.hist(randn(100), bins=20, color='k', alpha=0.3) # histogram, 20구간, 투명도 0.3
ax2.scatter(np.arange(30), np.arange(30)+3*randn(30)) # 점찍기 명시 

# ax3, ax1은 점 or 선 
# ax2 는 scatter : 명확하게 점을 찍겠다고 명시 
# _ : return 값을 나타내지 않겠다. (출력하지 않겠다.)
```

```python
# 그래프 Drawing 
plt.show() # 그래프 출력 안됨 
fig # 캔버스 호출 
```

```python
# 메모리 관리 차원 

- close() # closes the current figure
- close(fig) # closes the Figure instance fig
- close(num) # closes the figure number num
- close(name) # closes figure with that label name.
- close('all') # closes all the figure windows
```
