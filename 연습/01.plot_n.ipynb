#### [Quiz] iris 데이터를 불러와서
- (12, 12) 사이즈의 캔버스를 만들고
- (4, 4) 사이즈의 axes를 생성한 후,
- 각 x, y축을 공유한 후,
- 대각 위치의 도화지에는 각 변수들의 histogram을 그리고
- 다른 위치의 도화지에는 두 변수 간의 산점도를 그리자.

```python
from sklearn.datasets import load_iris
data = load_iris()
X, fn = data.data, data.feature_names # 각 변수(의 이름)
```

```python
fig, axes = plt.subplots(4,4, figsize=(12,12), sharex=True, sharey=True)

ws = np.ones(X.shape[0])*0.2    # from where? x값의 가중치데이터 (y축 값)
n, p = axes.shape

for i in range(n):
    for j in range(p):
        ax = axes[i,j]
        if i == j : 
            ax.hist(X[:,i], weights=ws)
        else :
            ax.plot(X[:,j],X[:,i],'ro', markersize=1)
        if j == 0:
            ax.set_ylabel(fn[i])    # Y
        if i == 3:
            ax.set_xlabel(fn[j])    # Y


plt.subplots_adjust(wspace=0, hspace=0)
#(4)6~ 
```
